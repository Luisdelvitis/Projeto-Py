```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo CSV
df = pd.read_csv('lista_de_espera_sisu_2022_2.csv', sep='|', encoding='latin-1')

# Filtrar os dados para o curso de Medicina
df_medicina = df[df['NOME_CURSO'].str.contains('MEDICINA', case=False, na=False)]

# Contar a quantidade de pessoas por estado
# Assumindo que 'UF_CANDIDATO' é o estado do candidato
quantidade_pessoas_estado_medicina = df_medicina['UF_CANDIDATO'].value_counts()

# Exibir a quantidade de pessoas por estado
print(quantidade_pessoas_estado_medicina)

# Plotar a quantidade de pessoas por estado
plt.figure(figsize=(12, 6))
plt.bar(quantidade_pessoas_estado_medicina.index, quantidade_pessoas_estado_medicina.values)
plt.xlabel('Estado')
plt.ylabel('Quantidade de Pessoas')
plt.title('Quantidade de Pessoas na Lista de Espera para Medicina por Estado')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from fpdf import FPDF

# Carregar os dados
df = pd.read_csv('lista_de_espera_sisu_2022_2.csv', sep='|', encoding='latin-1')

# Filtrar por curso de medicina
df_medicina = df[df['NOME_CURSO'].str.contains('MEDICINA', case=False, na=False)]

# Contar a quantidade de pessoas na lista de espera por estado
alunos_estado = df_medicina['UF_CANDIDATO'].value_counts()

# Criar o gráfico de distribuição de vagas por estado
plt.figure(figsize=(10, 6))
alunos_estado.plot(kind='bar', color='skyblue')
plt.title('Quantidade de Pessoas na Lista de Espera de Medicina por Estado')
plt.xlabel('Estado')
plt.ylabel('Quantidade de Pessoas')
plt.tight_layout()

# Salvar o gráfico como imagem
plt.savefig('lista_espera_medicina_estado.png')

# Criar o PDF
pdf = FPDF()

# Adicionar página
pdf.add_page()

# Adicionar título
pdf.set_font("Arial", 'B', 16)
pdf.cell(200, 10, txt="Relatório: Lista de Espera Medicina", ln=True, align='C')

# Adicionar o gráfico de vagas por estado
pdf.image('lista_espera_medicina_estado.png', x=10, y=30, w=190)

# Gerar o PDF
pdf.output("relatorio_medicina.pdf")

print("PDF criado com sucesso!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados
df = pd.read_csv('lista_de_espera_sisu_2022_2.csv', sep='|', encoding='latin-1')

# Filtrar por curso de medicina
df_medicina = df[df['NOME_CURSO'].str.contains('MEDICINA', case=False, na=False)]

# Contar a quantidade total de candidatos para todos os cursos e medicina
total_candidatos = df.shape[0]
total_medicina = df_medicina.shape[0]
outros_cursos = total_candidatos - total_medicina

# Criar gráfico de pizza
plt.figure(figsize=(8, 8))
labels = ['Medicina', 'Outros Cursos']
sizes = [total_medicina, outros_cursos]
colors = ['#ff9999', '#66b3ff']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Distribuição de Candidatos na Lista de Espera: Medicina vs Outros Cursos')
plt.axis('equal')  # Assegurar que o gráfico é desenhado como um círculo

# Exibir o gráfico
plt.show()

# Salvar o gráfico como imagem
plt.savefig('distribuicao_medicina_vs_outros.png')
```

```python
from fpdf import FPDF

# Criar a classe PDF
class PDF(FPDF):
    def header(self):
        pass  # Nenhum cabeçalho necessário
    
    def footer(self):
        # Rodapé opcional
        self.set_y(-15)
        self.set_font("Arial", "I", 8)
        self.cell(0, 10, f"Page {self.page_no()}", align="C")

# Inicializar o PDF
pdf = PDF()
pdf.set_left_margin(15)  # Margem esquerda
pdf.set_right_margin(15)  # Margem direita
pdf.set_top_margin(15)  # Margem superior
pdf.set_auto_page_break(auto=True, margin=15)  # Margem para quebra de página

# Adicionar capa
pdf.add_page()
pdf.image('capa.png', x=15, y=15, w=pdf.w - 30)
pdf.ln(85)  # Espaço após a capa

pdf.set_font("Arial", 'B', 16)
pdf.cell(0, 10, "Desigualdade no Acesso ao Curso de Medicina no Brasil", ln=True, align='C')
pdf.ln(5)

pdf.set_font("Arial", '', 12)
pdf.multi_cell(0, 10, "Apesar do sucesso das políticas afirmativas nos últimos anos, a medicina ainda é uma profissão dominada por brancos oriundos das classes média e alta. De acordo com o Instituto Brasileiro de Geografia e Estatística (IBGE), mais de 55% da população brasileira se identifica como preta (9,4%) ou parda (46,8%) e, apesar do quantitativo expressivo, quando o assunto é representatividade dessa população na medicina, as estatísticas encolhem.\n")

pdf.multi_cell(0, 10, "O curso de Medicina é, historicamente, um dos mais almejados e prestigiados no Brasil, tanto pela relevância social da profissão quanto pelas boas perspectivas salariais e de carreira. No entanto, o acesso a essas formações, sobretudo nas universidades federais, é marcado por um desequilíbrio significativo, refletindo a desigualdade estrutural do país.\n")

# Adicionar gráfico 1 (Distribuição Medicina vs Outros)
pdf.add_page()
pdf.image('distribuicao_medicina_vs_outros.png', x=15, y=15, w=pdf.w - 30)
pdf.ln(190)  # Espaço após a imagem

pdf.set_font("Arial", 'B', 16)
pdf.cell(0, 10, "A Relação Vagas x Candidatos Em Lista de Espera", ln=True, align='C')
pdf.ln(5)

pdf.set_font("Arial", '', 12)
pdf.multi_cell(0, 10, "Anualmente, milhares de jovens brasileiros competem por uma das raras vagas nos cursos de medicina das universidades federais. A oferta de vagas, no entanto, está muito aquém da demanda. Segundo dados recentes, algumas instituições chegam a ter uma relação de mais de 200 candidatos por vaga, criando um funil quase intransponível para muitos aspirantes.\nQuantidade de alunos na lista de espera para o curso de Medicina no SISU 2022/2: 45.508 candidatos.\n")

# Adicionar gráfico 2 (Lista de Espera Medicina por Estado)
pdf.add_page()
pdf.image('lista_espera_medicina_estado.png', x=15, y=15, w=pdf.w - 30)
pdf.ln(140)  # Espaço após a imagem

pdf.set_font("Arial", 'B', 16)
pdf.cell(0, 10, "Desigualdade no Acesso", ln=True, align='C')
pdf.ln(5)

pdf.set_font("Arial", '', 12)
pdf.multi_cell(0, 10, "Esse cenário evidencia as disparidades econômicas e sociais no Brasil. Embora o ensino superior nas universidades federais seja gratuito, a preparação para alcançar uma dessas vagas não o é. Alunos de escolas particulares, que contam com uma formação de qualidade superior à média das escolas públicas, têm muito mais chances de ingressar. Em contrapartida, estudantes de escolas públicas, que representam a maioria da população, enfrentam não apenas dificuldades educacionais, mas também a escassez de recursos financeiros para custear cursos preparatórios que os preparem de forma competitiva para o ENEM.\n")

# Adicionar imagem de desigualdade social
pdf.add_page()
pdf.image('desigualdades-sociais.jpeg', x=15, y=15, w=pdf.w - 30)
pdf.ln(80)  # Espaço após a imagem

pdf.set_font("Arial", 'B', 16)
pdf.cell(0, 10, "O Sistema de Cotas e Suas Limitações", ln=True, align='C')
pdf.ln(5)

pdf.set_font("Arial", '', 12)
pdf.multi_cell(0, 10, "O sistema de cotas, implementado como uma tentativa de mitigar essa desigualdade, tem proporcionado avanços significativos no acesso de alunos de escolas públicas, negros, pardos e indígenas ao ensino superior. No entanto, ele ainda enfrenta resistência em certos segmentos da sociedade e não resolve completamente o problema. Muitos candidatos cotistas, mesmo com as políticas afirmativas, continuam a enfrentar dificuldades para ingressar nos cursos mais concorridos, como Medicina.\nO número de vagas é limitado e a lista de espera continua longa, o que gera frustração para muitos desses jovens. Mesmo com as cotas, a quantidade de estudantes que conseguem entrar no curso de Medicina ainda está longe de refletir a realidade da população brasileira.\n")

# Adicionar Considerações Finais
pdf.add_page()
pdf.set_font("Arial", 'B', 16)
pdf.cell(0, 10, "Considerações Finais", ln=True, align='C')
pdf.ln(5)

pdf.set_font("Arial", '', 12)
pdf.multi_cell(0, 10, "A desigualdade no acesso ao curso de Medicina nas universidades federais brasileiras é uma manifestação de problemas maiores que permeiam o sistema educacional e social do país. Para mitigar essa realidade, seria necessária uma ampliação significativa da oferta de vagas nas universidades públicas, aliada a uma reforma no ensino básico, que melhore a qualidade da educação nas escolas públicas e ofereça reais condições de competitividade para todos os estudantes. O fortalecimento de políticas de permanência estudantil também seria essencial para garantir que aqueles que conseguem ingressar no curso de Medicina possam concluí-lo, independentemente de suas condições socioeconômicas.\n")

# Salvar o PDF
pdf.output("Relatorio_Desigualdade_Medicina.pdf")
```
